````python
# imports cell
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# normalize function
def normalize(x):
    # lookup dictionary
    lookup = {
        'asp': 'Aspen',
        'bar': 'Barranca',
        'cha': 'Chamisa',
    }

    name = str(x.School)
    x.School = lookup[name[0:3].lower()] if name[0:3].lower() in lookup else name
````

````python
df = pd.read_excel('eyh2020.xlsx', skiprows=13, index_col=0, usecols=[0,1])
df.dropna(inplace=True)
df.apply(normalize, axis=1)
count = df.groupby('School')['School'].count()
print(count)
````

````python
count.plot(kind='bar')
plt.show()
````
